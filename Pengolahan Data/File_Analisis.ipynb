### Cleaning Data

```python
#menghapus nilai null
df_airbnb.dropna(inplace=True)
```

```python
#Menghapus data pada kolom price yang bernilai 0
df_airbnb = df_airbnb.drop(df_airbnb[df_airbnb['price'] == 0].index)
```

## DATA MANIPULATION

### 1. Menampilkan harga penginapan termurah sampai termahal

```python
df_airbnb_sort = df_airbnb.sort_values(by=['price'], ascending= True)

df_airbnb_sort
```

### 2. Menampilkan list penginapan yang bertipe Private room dan memiliki harga di bawah 150

```python
df_airbnb_kisaran = df_airbnb.loc[(df_airbnb['room_type'] == 'Private room') & (df_airbnb['price'] <= 150)]

df_airbnb_kisaran
```

### 3. Menampilkan list 5 penginapan yang memiliki jumlah ketersediaan hari terbanyak dalam 1 tahun

```python
df_airbnb_review = df_airbnb
df_airbnb_review['availability_category'] = np.where(df_airbnb_review['availability_365'] > 200, 'tersedia_sangat_banyak', np.where(df_airbnb_review['availability_365'] > 50, 'tersedia_banyak', 'terbatas'))
df_airbnb_banyak_review = df_airbnb_review[df_airbnb_review['availability_category'] == 'tersedia_sangat_banyak']
df_airbnb_banyak_review = df_airbnb_banyak_review.sort_values('availability_category', ascending=False)

df_airbnb_banyak_review.head()
```

### 4. Menampilkan harga min, max dan rata_rata setiap type ruangan

```python
df_type = df_airbnb.groupby('room_type')['price'].agg(harga_terendah = 'min',
                                                      harga_tertinggi = 'max',
                                                      harga_rata_rata = 'mean')
df_type
```

### 5. Menampilkan list 5 host yang memiliki penginapan terbanyak

```python
df_jumlah_penginapan = df_airbnb.groupby('host_name')['name'].agg(jumlah_penginapan = 'count')

df_jumlah_penginapan.sort_values("jumlah_penginapan", ascending=False).head()
```

### 6. Menampilkan rata-rata harga penginapan di setiap wilayah

```python
rata_rata_harga = df_airbnb.groupby('neighbourhood')['price'].mean().reset_index(name='Harga_rata_rata')

rata_rata_harga
```

### 7. Menampilkan perkiraan pendapatan masing masing pemilik penginapan

```python
df_jumlah_penginapan = df_airbnb.groupby('host_name').agg(jumlah_penginapan=('name', 'count'),
                                                          rata_rata_harga_penginapan=('price', 'mean'))

df_jumlah_penginapan['jumlah_penghasilan'] = df_jumlah_penginapan['jumlah_penginapan'] * df_jumlah_penginapan['rata_rata_harga_penginapan']

df_jumlah_penginapan.sort_values('jumlah_penginapan', ascending=False)
```

### 8.Menampilkan jumlah review terbanyak dari setiap penginapan pada tanggal tertentu

```python
df_hour = df_airbnb.groupby(['last_review', 'name'])
df_waktu = df_hour['number_of_reviews'].agg(jumlah_review='sum')

df_waktu.sort_values('jumlah_review', ascending=False).head(20)
```

## DATA VISUALISATION

## Diagram Batang

### Mengelompokkan data berdasarkan kolom 'host_name' dalam dataset Airbnb dan menghitung jumlah penginapan per tuan rumah

```python
# Mengelompokkan data berdasarkan 'host_name' dan menghitung jumlah penginapan per tuan rumah
listing_count = df_airbnb['host_name'].value_counts()

# Mengambil 10 tuan rumah dengan jumlah penginapan terbanyak
top_10_hosts = listing_count.head(10)

# Membuat grafik batang
plt.bar(top_10_hosts.index, top_10_hosts.values)

# Mengatur label sumbu x dan y serta judul
plt.xlabel('Host Name')
plt.ylabel('Jumlah Penginapan')
plt.title('Jumlah Daftar Penginapan berdasarkan pemilik')

# Memutar teks pada label sumbu x agar lebih mudah dibaca
plt.xticks(rotation=90)

# Menampilkan grafik batang
plt.show()
```

### mengelompokkan data berdasarkan kolom 'room_type' dalam dataset Airbnb dan menghitung jumlah entri dalam setiap kategori

```python
# Menghitung rata-rata harga penginapan per neighbourhood_group
average_price = df_airbnb.groupby('neighbourhood_group')['price'].mean()

# Membuat grafik batang
plt.bar(average_price.index, average_price.values)

# Mengatur label sumbu x dan y serta judul
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.title('Rata-rata harga penginapan pada setiap kelompok wilayah')

# Memutar teks pada label sumbu x agar lebih mudah dibaca
plt.xticks(rotation=90)

# Menampilkan grafik batang
plt.show()
```

## Pie Chart

### Mengelompokkan data berdasarkan kolom 'room_type' dan menghitung jumlah entri dalam setiap kategori

```python
# Mengelompokkan data berdasarkan 'room_type' dan menghitung jumlah entri dalam setiap kategori
room_type_counts = df_airbnb['room_type'].value_counts()

# Membuat pie chart
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%')

# Menambahkan judul
plt.title('Distribution Room_type')

# Menampilkan pie chart
plt.show()
```

## Line Chart

### Menampilkan Data Jumlah Review dari tahun 2011 sampai 2019

```python
# Mengonversi kolom 'last_review' ke tipe data datetime
df_airbnb['last_review'] = pd.to_datetime(df_airbnb['last_review'])

# Mengurutkan data berdasarkan kolom 'last_review'
df_airbnb.sort_values(by='last_review', inplace=True)

# Mengelompokkan data berdasarkan tahun dan menghitung jumlah review
review_count_by_year = df_airbnb['last_review'].dt.year.value_counts().sort_index()

# Membuat grafik garis
plt.plot(review_count_by_year.index, review_count_by_year.values, marker='o', linestyle='-')

# Mengatur label sumbu x dan y serta judul
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.title('Review Count Over Time')

# Menampilkan grafik garis
plt.show()
```
